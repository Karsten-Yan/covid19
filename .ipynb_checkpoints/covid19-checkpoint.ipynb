# Module import

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go

sns.set(style="whitegrid", rc={"lines.linewidth": 5}, palette="muted")
```

# Data mining

```python
df = pd.read_csv("https://raw.githubusercontent.com/datasets/covid-19/master/time-series-19-covid-combined.csv", parse_dates=[4])
```

# Data Cleaning

```python
cols = ["state","country","lat","long","date","confirmed","recovered","deaths"]
df.columns = cols
```

```python
df.country = df.country.replace(to_replace = " ", value = "_", regex = True).str.lower()
df.state = df.state.replace(to_replace = " ", value = "_", regex = True).str.lower()
df.head()
```

```python
df.info()
```

```python
df2=df[df["country"] == "mainland_china"].groupby("date").sum()
df2.reset_index(inplace=True)
df3=df.groupby("date").sum()
df3.reset_index(inplace=True)
```

```python
fig, ax = plt.subplots(3, 3, figsize=(20,10),sharex="col",
                       #sharey="row",
                       gridspec_kw={'hspace': 0})
fig.suptitle('Covid-19', fontsize=30)

sns.lineplot(x="date", y="confirmed", data=df[df["country"] == "germany"],  ax=ax[0][0])
sns.lineplot(x="date", y="recovered", data=df[df["country"] == "germany"],  ax=ax[1][0])
sns.lineplot(x="date", y="deaths", data=df[df["country"] == "germany"], ax=ax[2][0])
ax[0][0].tick_params(labelrotation=45, labelsize=12)
ax[0][0].set_ylabel("confirmed",fontsize=20)
ax[0][0].set_title("Germany",fontsize=20)
ax[1][0].tick_params(labelrotation=45, labelsize=12)
ax[1][0].set_ylabel("recovered",fontsize=20)
ax[2][0].tick_params(labelrotation=45, labelsize=12)
ax[2][0].set_xlabel("date",fontsize=20)
ax[2][0].set_ylabel("deaths",fontsize=20)

sns.lineplot(x="date", y="confirmed", data=df2,  ax=ax[0][1])
sns.lineplot(x="date", y="recovered", data=df2,  ax=ax[1][1])
sns.lineplot(x="date", y="deaths", data=df2, ax=ax[2][1])
ax[0][1].tick_params(labelrotation=45, labelsize=12)
ax[0][1].set_title("Mainland China",fontsize=20)
ax[0][1].set_ylabel("",fontsize=20)
ax[1][1].tick_params(labelrotation=45, labelsize=12)
ax[1][1].set_ylabel("",fontsize=20)
ax[2][1].tick_params(labelrotation=45, labelsize=12)
ax[2][1].set_xlabel("date",fontsize=20)
ax[2][1].set_ylabel("",fontsize=20)


sns.lineplot(x="date", y="confirmed", data=df3,  ax=ax[0][2])
sns.lineplot(x="date", y="recovered", data=df3,  ax=ax[1][2])
sns.lineplot(x="date", y="deaths", data=df3, ax=ax[2][2])
ax[0][2].tick_params(labelrotation=45, labelsize=12)
ax[0][2].set_ylabel("",fontsize=20)
ax[0][2].set_title("World",fontsize=20)
ax[1][2].tick_params(labelrotation=45, labelsize=12)
ax[1][2].set_ylabel("",fontsize=20)
ax[2][2].tick_params(labelrotation=45, labelsize=12)
ax[2][2].set_xlabel("date",fontsize=20)
ax[2][2].set_ylabel("",fontsize=20)
;
```

```python
df_group = df.groupby(["lat","long"]).sum()
df_group.reset_index(inplace=True)
df_group.sort_values("confirmed", ascending = False).head()
```

```python
plt.figure(figsize=(16, 16))
sns.scatterplot(x=df_group.long, y=df_group.lat,hue=df_group.confirmed, size=df_group.recovered,sizes=(40, 400))
```
