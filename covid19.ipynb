```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
```

```python
df = pd.read_csv("time-series-19-covid-combined.csv", parse_dates=[4])
```

```python
cols = ["state","country","lat","long","date","confirmed","recovered","deaths"]
df.columns = cols
df.head()
```

```python
df.info()
```

```python
#df.drop("state", inplace = True, axis =1)
```

```python
df_new=df[df["country"] == "Mainland China"].drop_duplicates("date")
```

```python
fig, ax = plt.subplots(3, 1, figsize=(10,10),sharex=True,gridspec_kw={'hspace': 0})
fig.suptitle('Covid 19 in Germany', fontsize=30)
sns.lineplot(x="date", y="confirmed", data=df[df["country"] == "Germany"],  ax=ax[0])
sns.lineplot(x="date", y="recovered", data=df[df["country"] == "Germany"],  ax=ax[1])
sns.lineplot(x="date", y="deaths", data=df[df["country"] == "Germany"], ax=ax[2])
ax[0].tick_params(labelrotation=45, labelsize=12)
ax[0].set_xlabel("date",fontsize=20)
ax[0].set_ylabel("confirmed",fontsize=20)
ax[1].tick_params(labelrotation=45, labelsize=12)
ax[1].set_xlabel("date",fontsize=20)
ax[1].set_ylabel("recovered",fontsize=20)
ax[2].tick_params(labelrotation=45, labelsize=12)
ax[2].set_xlabel("date",fontsize=20)
ax[2].set_ylabel("deaths",fontsize=20)
;
```

```python
df2=df[df["country"] == "Mainland China"].groupby("date").sum()
df2.reset_index(inplace=True)
```

```python
fig, ax = plt.subplots(3, 1, figsize=(10,10),sharex=True,gridspec_kw={'hspace': 0})
fig.suptitle('Covid 19 in China', fontsize=30)
sns.lineplot(x="date", y="confirmed", data=df2,  ax=ax[0])
sns.lineplot(x="date", y="recovered", data=df2,  ax=ax[1])
sns.lineplot(x="date", y="deaths", data=df2, ax=ax[2])
ax[0].tick_params(labelrotation=45, labelsize=12)
ax[0].set_xlabel("date",fontsize=20)
ax[0].set_ylabel("confirmed",fontsize=20)
ax[1].tick_params(labelrotation=45, labelsize=12)
ax[1].set_xlabel("date",fontsize=20)
ax[1].set_ylabel("recovered",fontsize=20)
ax[2].tick_params(labelrotation=45, labelsize=12)
ax[2].set_xlabel("date",fontsize=20)
ax[2].set_ylabel("deaths",fontsize=20)
;
```

```python
df3=df.groupby("date").sum()
df3.reset_index(inplace=True)
```

```python
fig, ax = plt.subplots(3, 1, figsize=(10,10),sharex=True,gridspec_kw={'hspace': 0})
fig.suptitle('Covid 19 in World', fontsize=30)
sns.lineplot(x="date", y="confirmed", data=df3,  ax=ax[0])
sns.lineplot(x="date", y="recovered", data=df3,  ax=ax[1])
sns.lineplot(x="date", y="deaths", data=df3, ax=ax[2])
ax[0].tick_params(labelrotation=45, labelsize=12)
ax[0].set_xlabel("date",fontsize=20)
ax[0].set_ylabel("confirmed",fontsize=20)
ax[1].tick_params(labelrotation=45, labelsize=12)
ax[1].set_xlabel("date",fontsize=20)
ax[1].set_ylabel("recovered",fontsize=20)
ax[2].tick_params(labelrotation=45, labelsize=12)
ax[2].set_xlabel("date",fontsize=20)
ax[2].set_ylabel("deaths",fontsize=20)
;
```
